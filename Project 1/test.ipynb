```python
import pandas as pd
```

```python
df=pd.read_excel("2023-R1.xlsx")
df
```

```python
# Get unique course code and course name pairs
unique_courses = df[['Course Code', 'Course Name']].drop_duplicates()

# Convert to dictionary with course code as key and course name as value
course_dict = dict(zip(unique_courses['Course Code'], unique_courses['Course Name']))
course_dict = dict(sorted(course_dict.items()))
for i,j in course_dict.items():
    print(i," : ",j)
```
